```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data
variables = ['Male', 'Dark', 'Light', 'Medium', 'Non-human', 'Sentiment Score']
year_2017_2021 = [-0.3059, -0.2934, -0.1827, -0.2497, 1.4223, 2.0334]
year_2017_2022 = [-0.3223, -0.3728, -0.3216, -0.4176, 1.3680, 1.8304]
year_2021_2022 = [0.0532, -0.1063, -0.0119, -0.0506, 3.1484, 0.3885]

# Set up the plot
fig, ax = plt.subplots()
ax.set_xlabel('Coefficient')
ax.set_ylabel('Variable')
ax.set_title('Lollipop Plot')

# Plot the data
for i, var in enumerate(variables):
    y = [i, i, i]  # Repeat the variable three times to create the lollipop effect
    x = [year_2017_2021[i], year_2017_2022[i], year_2021_2022[i]]
    colors = ['tab:blue', 'tab:orange', 'tab:green']  # Set different colors for the coefficients
    for j in range(2):
        ax.plot([x[j], x[j+1]], [i, i], color=colors[j], linewidth=1, alpha=0.7)  # Add lines between the coefficients
    ax.scatter(x, y, color=colors, alpha=0.7)  # Dots

# Add a legend
ax.legend(['Regression 1', 'Regression 2', 'Regression 3'])

# Set the y-axis labels to the variable names
ax.set_yticks(np.arange(len(variables)))
ax.set_yticklabels(variables)

# Add legend for each color of dots
blue_dot = plt.Line2D((0,1),(0,0), color='tab:blue', marker='o', linestyle='')
orange_dot = plt.Line2D((0,1),(0,0), color='tab:orange', marker='o', linestyle='')
green_dot = plt.Line2D((0,1),(0,0), color='tab:green', marker='o', linestyle='')
ax.legend([blue_dot, orange_dot, green_dot], ['2017-2021', '2017-2022', '2021-2022'], loc='lower right')

# Show the plot
plt.show()
```

```python
# Define the data
variables = ['Male', 'Dark', 'Light', 'Medium', 'Non-human', 'Sentiment Score']
year_2017_2021 = [-0.3059, -0.2934, -0.1827, -0.2497, 1.4223, 2.0334]
year_2017_2022 = [-0.3223, -0.3728, -0.3216, -0.4176, 1.3680, 1.8304]
year_2021_2022 = [0.0532, -0.1063, -0.0119, -0.0506, 3.1484, 0.3885]
sig_2017_2021 = [3, 3, 2, 3, 3, 3]
sig_2017_2022 = [3, 3, 3, 3, 3, 3]
sig_2021_2022 = [3, 3, 0, 1, 3, 2]

# Set up the plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlabel('Coefficient', fontsize=12)
ax.set_ylabel('Variable', fontsize=12)
ax.set_title('The Comparison of the Performance of Price Predictive Variables (Before and After 2021)', fontsize=14)

# Plot the data
for i, var in enumerate(variables):
    y = [i, i, i]  # Repeat the variable three times to create the lollipop effect
    x = [year_2017_2021[i], year_2017_2022[i], year_2021_2022[i]]
    colors = ['tab:blue', 'tab:orange', 'tab:green']  # Set different colors for the coefficients
    for j in range(2):
        ax.plot([x[j], x[j+1]], [i, i], color='gray', linewidth=2, alpha=1)  # Add lines between the coefficients
    for j, sig in enumerate([sig_2017_2021[i], sig_2017_2022[i], sig_2021_2022[i]]):
        if j == 0:
            ax.scatter(x[j], y[j], color=colors[j], s=250, alpha=1, label='Regression {}'.format(j+1))
        else:
            ax.scatter(x[j], y[j], color=colors[j], s=250, alpha=1)
        ax.text(x[j], y[j], sig, color='white', ha='center', va='center', fontweight='bold', fontsize=12)  # Add significance level to the dots

# Set the y-axis labels to the variable names
ax.set_yticks(np.arange(len(variables)))
ax.set_yticklabels(variables, fontsize=12)

# Add legend for each color of dots
blue_dot = plt.Line2D((0,1),(0,0), color='tab:blue', marker='o', linestyle='')
orange_dot = plt.Line2D((0,1),(0,0), color='tab:orange', marker='o', linestyle='')
green_dot = plt.Line2D((0,1),(0,0), color='tab:green', marker='o', linestyle='')

sig_legend = {0: 'Not significant', 1: '*', 2: '**', 3: '***'}
handles = [plt.Line2D([0], [0], marker='$3$', color='grey', label=sig_legend[i], markerfacecolor='k') for i in range(4)]

ax.legend([blue_dot, orange_dot, green_dot] + handles, ['2017-2021', '2017-2022', '2021-2022', 'Significance Level'], loc='lower right', fontsize=12)

# Show the plot
plt.show()
```

```python
# Define the data
variables = ['Male', 'Dark', 'Light', 'Medium', 'Non-human', 'Sentiment Score']
non_sentiment = [-0.2906, -0.2735, -0.1484, -0.2302, 1.4861, 0]
sentiment = [-0.3059, -0.2934, -0.1827, -0.2497, 1.4223, 2.0334]
sig_no = [3, 3, 1, 3, 3, 0]
sig_yes = [3, 3, 2, 3, 3, 3]

# Set up the plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlabel('Coefficient', fontsize=12)
ax.set_ylabel('Variable', fontsize=12)
ax.set_title('The Comparison of the Performance of Price Predictive Variables (With/Without Sentiment Score)', fontsize=14)

# Plot the data
for i, var in enumerate(variables):
    y = [i, i]  # Repeat the variable two times to create the lollipop effect
    x = [non_sentiment[i], sentiment[i]]
    colors = ['tab:blue', 'tab:green']  # Set different colors for the coefficients
    if i == 5:
        ax.scatter(x[1], y[1], color=colors[1], s=250, alpha=1, label='Regression {}'.format(2))
        ax.text(x[1], y[1], sig_yes[i], color='white', ha='center', va='center', fontweight='bold', fontsize=12)
    else:
        for j in range(1):
            ax.plot([x[j], x[j+1]], [i, i], color='gray', linewidth=2, alpha=1)  # Add lines between the coefficients
        for j, sig in enumerate([sig_no[i], sig_yes[i]]):
            if j == 0:
                ax.scatter(x[j], y[j], color=colors[j], s=250, alpha=1, label='Regression {}'.format(j+1))
            else:
                ax.scatter(x[j], y[j], color=colors[j], s=250, alpha=1)
            ax.text(x[j], y[j], sig, color='white', ha='center', va='center', fontweight='bold', fontsize=12)  # Add significance level to the dots

# Set the y-axis labels to the variable names
ax.set_yticks(np.arange(len(variables)))
ax.set_yticklabels(variables, fontsize=12)

# Add legend for each color of dots
blue_dot = plt.Line2D((0,1),(0,0), color='tab:blue', marker='o', linestyle='')
green_dot = plt.Line2D((0,1),(0,0), color='tab:green', marker='o', linestyle='')

sig_legend = {0: 'Not significant', 1: '*', 2: '**', 3: '***'}
handles = [plt.Line2D([0], [0], marker='$3$', color='grey', label=sig_legend[i], markerfacecolor='k') for i in range(4)]

ax.legend([blue_dot, green_dot] + handles, ['With Sentiment Score', 'Without Sentiment Score', 'Significance Level'], loc='lower right', fontsize=12)

# Show the plot
plt.show
```
